```python
import pandas as pd
```

```python
# Read data from CSV file
df = pd.read_csv('Inputdata - Sheet1.csv')
df
```

```python
# Fill missing values with zeros
df.fillna(0, inplace=True)
```

```python
# Convert 'Day' column to datetime format
df['Day'] = pd.to_datetime(df['Day'])
```

```python
# Drop unnecessary columns
df.drop(['Reporting starts', 'Reporting ends'], axis=1, inplace=True)
```

```python
# Create new columns
df['Ad spent'] = df['Amount spent (INR)']
df['No.of Leads'] = df['Results']
df['Cost per Leads'] = df['Ad spent'] / df['No.of Leads']
df['CTR'] = df['CTR (link click-through rate)']
df['LP views'] = df['Landing page views']
df['Member passed'] = (df['Landing page views'] / df['Link clicks']) * 100
df['LP Conversion%'] = (df['No.of Leads'] / df['Landing page views']) * 100
```

```python
df['LP Conversion%']  =df['LP Conversion%'] .map(lambda x: f'{x:.2f}%')
```

```python
df['Member passed'] = df['Member passed'].map(lambda x: f'{x:.2f}%')
```

```python
df['CTR']=df['CTR'].map(lambda x: f'{x:.2f}%')
```

```python
# Create 'Date' and 'LP Views' columns
df['Date'] = df['Day']
df['LP Views'] = df['Landing page views']
```

```python
output_df = df[['Day', 'Campaign name', 'Amount spent (INR)', 'Results', 'Cost per Leads',
                'CTR', 'Link clicks', 'Landing page views', 'Member passed', 'LP Conversion%']]
```

```python
# Rename columns
output_df.columns = ['Date', 'Campaign Name', 'Ad spent', 'No.of Leads', 'Cost Per Lead',
                     'CTR', 'Link Clicks', 'LP Views', 'Member Passed', 'LP Conversion %']
```

```python
# Sort the DataFrame by the 'Date' column in ascending order
output_df = output_df.sort_values(by='Date')

# Reset the index after sorting
output_df = output_df.reset_index(drop=True)
```
